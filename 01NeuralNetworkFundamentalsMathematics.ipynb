# 일차함수
y = ax + b

```python
def linear_function(x):
    a = 2.5
    b = 4
    return a*x + b
```

```python
print(linear_function(5)) #5*2.5+4
```

```python
x = np.arange(-5, 5, 0.1)
y = linear_function(x)

plt.title("linear funtion(2.5x + 4)")
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## 이차함수
y = ax²+ bx + c

```python
def quadratic_function(x):
    a = 1
    b = -1
    c = -2
    return a*x*x + b*x +c
```

```python
print(quadratic_function(5)) # 1*5*5 + (-1)*5 + (-2)
```

```python
x = np.arange(-5, 5, 0.1)
y = quadratic_function(x)

plt.title("quadratic function(x^2-x-2)")
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## 삼차함수(다항함수)
y = ax³ + bx² + cx + d

```python
def cubic_function(x):
    a = 3
    b = 0
    c = -2
    d = 5
    return a*x**3 + b*x**2 +c*x +d
```

```python
print(cubic_function(3)) # 3*3^3+0*3^2+(-2)*3+5
```

```python
x = np.arange(-5, 5, 0.1)
y = cubic_function(x)

plt.title("cubic function(3x^2-2x+5)")
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## 함수의 최소값/최대값

```python
def myfunc(x):
    a = 1
    b = -3
    c = 10
    return a*x**2 + b*x + c
```

```python
x = np.arange(-10, 10, 0.1)
y = myfunc(x)

plt.title("myfunc(x^2-3x+10)")
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(1.5, myfunc(1.5))
plt.text(1.5-1.5, myfunc(1.5)+10, "min value of f(x)\n({},{})".format(1.5, myfunc(1.5)), fontsize=10)
plt.show()
```

```python
min_val = min(y)
print(min_val)
```

## 특정 구간 내에서 최솟값 구하기

```python
def get_min(x1, x2, f):
    x = np.arange(x1, x2, 0.01)
    y = f(x)
    
    plt.plot(x, y)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("get minimum")
    plt.show()
    
    return min(y)
```

```python
print(get_min(1,4, myfunc))
```

## 미분
어떤 한 순간의 변화량을 표시한 것

```python
''' 
파이썬에서 10e-50을 0.0으로 인식
그래서 1e-4정도로 하기
'''

def numerical_differential(f,x):
    h = 1e-4
    return ((f(x+h)-f(x-h))/(2*h))
```

```python
def my_func(x):
    return 2*x**2 + 3*x +2
```

```python
def linear_func(a,b,c,x): #a: x좌표 , b: y좌표, c:기울기
    return c*(x-a)+b
```

```python
c = numerical_differential(my_func, 1)
x = np.arange(-5, 5, 0.01)
y = linear_func(1, my_func(1), c, x)

plt.xlabel("x")
plt.ylabel("y")
plt.scatter(1, my_func(1)) #(1,7)에서 기울기
plt.plot(x, my_func(x), x, y, "r-")
plt.title("f(x)& linear function")
plt.show()
```

## 기울기(gradient)

```python
def numerical_diff(f, x):
    h = 1e-5
    grad = np.zeros_like(x)

    for i in range(x.size):
        tmp = x[i]
        x[i] = tmp+h
        fxh1 = f(x)
        
        x[i] = tmp-h
        fxh2 = f(x)
        
        grad[i] = (fxh1-fxh2)/(2*h)
        x[i] = tmp
    
    return grad
```

```python
print(numerical_diff(my_fucn2, np.array([3.0, 4])))
print(numerical_diff(my_fucn2, np.array([1.0, 2])))
```

## 기울기의 의미를 그래프로 확인
